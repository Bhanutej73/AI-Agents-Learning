Message Prompts<br>

Alternatively, you can pass in a list of messages to the model by providing a list of message objects.<br>

Message types:<br>

1. System message - Tells the model how to behave and provide context for interactions <br>

2. Human message - Represents user input and interactions with the model <br>

3. AI message - Responses generated by the model, including text content, tool calls, and metadata <br>

4. Tool message - Represents the outputs of tool calls <br>

System Message:<br>
A System Message represents an initial set of instructions that primes the model's behavior. You can use a system message to set the tone, define the model's role, and establish guidelines for responses.<br>

Human Message:<br>
A HumanMessage represents user input and interactions. They can contain text, images, audio, files, and any other amount of multimodal content.<br>

AI Message:<br>
An AlMessage represents the output of a model invocation. They can include multimodal data, tool calls, and provider-specific metadata that you can later access.<br>

Tool Message:<br>
For models that support tool calling, AI Messages can contain tool calls. Tool messages are used to pass the results of a single tool execution back to the model

Text Prompts:<br>
Text prompts are strings - ideal for straightforward generation tasks where you don't need to retain conversation history

```python
model.invoke("What is langchain?")
```

Messages:<br>
Messages are the fundamental unit of context for models in Langchain. They represent the input and output of models, carrying both the content and metadata needed to represent the state of a conversation when interacting with an LLM. Messages are objects that contain:<br>
1. Role-Identifies the message type(e.g. system,user)<br>
2. Content-Prepresents the actual content of the message IDs, and token usage<br>
3. MetaData-Optonal field such as respone information, message IDs, and token usage<br>

Langchain provides a standard message type that works across all model provisers, ensuring consistent behaviour regardless of the model being called.

```python
from langchain_core.messages import AIMessage, SystemMessage, HumanMessage,ToolMessage
messages=[
    SystemMessage(content="You are a poetry expert."),
    HumanMessage(content="What a poem in a special forces soldier's life?")
]

response=model.invoke(messages)
response.content
```

```python
system_msg=SystemMessage(
    """ You are a senior Python developer with expertise in web frameworks.
    Always provide code examples and explain your reasoning.
    Be concise but thorough in your responses."""
)

messages=[
    system_msg,
    HumanMessage("How do I create a REST API?")
]

response=model.invoke(messages)
print(response.content)
```

```python
human_msg=HumanMessage(
    content="Hello!",
    name="alice",
    id="12345"
)
```

```python
response=model.invoke([human_msg])
response
```

```python
ai_msg=AIMessage("I'd BE HAPPY TO HELP YOU WITH THAT QUESTION!")

messages=[
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Great! What's 2+2?")
]

response=model.invoke(messages)
print(response.content)
```

```python
response.usage_metadata
```

```python
#After a model makes a tool call
#Here, we demonstrate amnulayy creating the messages for brevity
ai_message=AIMessage(
    content=[],
    tool_calls=[{
        "name": "get_weather",
        "args": {"location": "New York"},
        "id": "call_123"
    }]
)

#Execute tool and create result message
weather_result="Sunny, 72 degrees Fahrenheit"
tool_message=ToolMessage(
    content=weather_result,
    tool_call_id="call_123"
)

messages=[
    HumanMessage("What's the weather in New York?"),
    ai_message, #Model's tool call
    tool_message #Tool execution result
]
responses=model.invoke(messages) #Model processes the results and generates final response
```
