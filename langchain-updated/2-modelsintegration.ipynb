Streaming:

Most of the models can stream their output content while it is being generated. 

```python
model.stream("Write me a 200 words paragraph on aritificial intelligence and its impact on society.")
```

```python
for chunk in model.stream("Write me a 200 words paragraph on why everyone should have more interaction with nature."):
    print(chunk.text, end="|", flush=True)
```

Batch:
<br>
Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel:

```python
responses=model.batch([
    "Why do parrots have colorful feathers?",
    "How do aeroplanes fly?",
    "What is quantum computing?"
    ]
)

for response in responses:
    print(response.content)
```

```python
responses=model.batch([
    "Why do parrots have colorful feathers?",
    "How do aeroplanes fly?",
    "What is quantum computing?"
    ],
    config={
        'max_concurrency':5
    }
)
```
