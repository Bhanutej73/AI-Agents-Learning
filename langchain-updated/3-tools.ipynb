<h1>Tools<h1>
<h4>Models can request to call tools that perform tasks such as fetching data from a database, searching the web or running code. Tools are pairings of:
<li>1. A schema, including the name of the tool, a description, and/or argument definitions (often a JSON schema)</li>
<li>2. A function or coroutine to execute.</li>
</h4>

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
from langchain.tools import tool
from langchain_core.messages import ToolMessage

@tool
def get_weather(location:str)->str:
    """Get the current weather of a location"""
    return "The current weather in {} is sunny with a temperature of 25 degrees Celsius."

model_with_tools=model.bind_tools([get_weather])
```

```python
response=model_with_tools.invoke("What is the current weather in New York?")
print(response)
for tool_call in response.tool_calls:
    #View the tool calls made by the model
    print(f"Tool:{tool_call['name']}")
    print(f"Args:{tool_call['args']}")
```

<h2>Tool Execution loops

```python
#Step-1: Model generates tool calls
messages=[{"role": "user", "content": "What is the current weather in New York?"}]
ai_msg=model_with_tools.invoke(messages)
messages.append(ai_msg)

#Step-2: Execute tools and collect results
for tool_call in ai_msg.tool_calls:
    #Execute the tool with the generated arguments
    tool_result=get_weather.invoke(tool_call)
    messages.append(
        ToolMessage(
            content=tool_result,
            tool_call_id=tool_call["id"]
        )
    )

#Step-3: Send tool results back to model for final response
final_response=model_with_tools.invoke(messages)
print(final_response.text)
```
