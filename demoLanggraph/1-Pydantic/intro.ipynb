```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name: str
    age: int
    city: str

person = Person(name="Alice", age=30, city="New York")
print(person)   
```

```python
type(person)
```

```python
from typing import Optional
class Employee(BaseModel):
    id:int
    name:str
    department: str
    salary:Optional[float] = None #Optional field with default value None
    is_active: Optional[bool] = True #Optional field with default value True
```

```python
emp1=Employee(id=1, name="Bob", department="HR")
print(emp1)
```

```python
from typing import List
class Classroom(BaseModel):
    room_number: str
    students: List[str] #List of strings
    capacity: int
```

```python
classroom =Classroom(room_number="A101", students=["Alice", "Bob", "Charlie"], capacity=30)
print(classroom)
```

```python
class Address(BaseModel):
    street: str
    city: str
    zip_code: str

class Customer(BaseModel):
    customer_id:int
    name: str
    address: Address

customer=Customer(customer_id=1,name="Emma",address={"street":"123 Main St", "city":"Springfield", "zip_code":"12345"})
print(customer)    
```

```python
#Pydantic Fields
from pydantic import Field

class Item(BaseModel):
    name:str=Field(min_length=3, max_length=50) #String field with length constraints
    price: float = Field(gt=0, le=1000) #Float field that must be greater than 0 and less than or equal to 1000
    quantity: int = Field(ge=0) #Integer field that must be greater than or equal to 0

#Valid instance
item1 = Item(name="Laptop", price=999.99, quantity=5)
print(item1)
```

```python
class User(BaseModel):
    username: str = Field(..., description="The unique username for the user") #Required field with description
    age:int = Field(default=18, description="The age of the user, default is 18") #Optional field with default value and description
    email: str = Field(default_factory=lambda: "default@example.com", description="The default email address of the user") #Optional field with default value and description

user1 = User(username="john_doe")
print(user1)

user2=User(username="jane_doe", age=25, email="jane@example.com")
print(user2)
```
